```python
import pandas as pd
```

```python
x=pd.read_csv("IBM_survey_scores.csv")
```

```python
x
```

```python
x.rename(columns={"Query3":"Query 3"},inplace=True)
```

```python
s='Query '
means=[]
for k in range(1,18):
    print("Score counts of "+s+str(k))
    for i,j in x[s+str(k)].value_counts().items(): print(i," occurs ",j," times")
    for i,j in x[s+str(k)].describe().items(): print(i," : ",j)
    means.append(x[s+str(k)].mean())
    print()
```

```python
print("Mean of mean scores of all queries : ",sum(means)/len(means))
```

```python
x["Average_Score"]=x.mean(axis=1)
```
