```python
import pandas as pd
import numpy as np
import re
import json
import os
```

```python
def create_schema(fname):
    print(fname)
    try:
        data = pd.read_csv(fname)
        table_object={"tablename": fname, "schema":{}}
        x=data.dtypes
        y=data.columns
        z=zip(y,x)
        for a,b in z:
            table_object["schema"][a]=str(b)
        f = open('all_schemas.txt','a')
        jsonstr = json.dumps(table_object)
        f.write(jsonstr+"\n")
        global count
        count+=1
    except pd.errors.ParserError:
        with open("errors.txt",'a') as error_file:
            print(f"Filename: {fname}", file=error_file)
```

```python
global count
count=0
for root, dirs, files in os.walk('.'):
   for fname in files:
       create_schema(fname) 
```
